```python
#Importing Libraries

import pandas as pd

import matplotlib.pyplot as plt

%matplotlib inline

from wordcloud import WordCloud

```

```python
#Importing Dataset

df=pd.read_csv('/content/abusive-ones.csv' , error_bad_lines=False , delimiter=";", usecols=["q_a"])
df.head(5)
```

```python
df1=df['q_a']
df1.head()
```

```python
text1 = " ".join(i for i in df.q_a)
```

```python
word_cloud1 = WordCloud(collocations = False, background_color = 'white').generate(text1)
```

```python
# Display the generated Word Cloud

plt.imshow(word_cloud1, interpolation='None')
```
